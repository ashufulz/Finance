#### Class 0 means the transaction is legit/normal, while Class 1 means it is fraud. Let's see the distribution of both classes

```python
main_df.Class.value_counts()
```

#### So as seen, this is an imbalanced dataset where Normal transactions are way more than Fraudulent ones. Hence if model is trained on this dataset, the results will be more biased on one side which is not appropriate!

### Separating data for analysis

```python
legit = main_df[main_df.Class==0]
fraud = main_df[main_df.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

### Statistical measures of the data

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

### Comparing values for both transactions

```python
main_df.groupby('Class').mean()
```

# UNDER_SAMPLING

#### Here we have around 3 Lac datapoints of Normal / Legit transactions whereas only 492 data points of Fraudulent ones. So we'll create a sample of legit dataset containing only 492 data points.

```python
legit_sample = legit.sample(n=492)
```

###  Concatenating both similar dataframes

```python
new_df = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_df.tail()
```

```python
new_df.groupby('Class').mean()
```

### Splitting data into Features and Target

```python
X = new_df.drop(columns='Class', axis=1)
y = new_df.Class
```

### Performing Train-Test split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=2)
```

# MODEL TRAINING

## Logistic Regression

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

### Accuracy Score

```python
y_train_pred = model.predict(X_train)
train_acc = accuracy_score(y_train_pred, y_train)
print('Accuracy of training data: ', train_acc)
```

```python
y_test_pred = model.predict(X_test)
test_acc = accuracy_score(y_test_pred, y_test)
print('Accuracy of test data: ', test_acc)
```
